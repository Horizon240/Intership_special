```python
# write function here
def c_to_f(x):
    return (x * 9/5) + 32
    




```

## Groupby

```python
# Create a dataframe using dictionary of your choice

# data = 
data = {
    'Food' : ['Fish','Chicken','Turkey','Shrimp'],
    'Likeness' : [3,9,8,6]
}
df = pd.DataFrame(data)
df
```

```python
# Use Groupby of single column with aggregate sum()
print(df.groupby('Food')['Likeness'].sum())

```

```python
# Use Groupby of single column with aggregate count()
print(df.groupby('Food')['Likeness'].count())
```

```python
# Use Groupby of single column with aggregate min() and max()
print(df.groupby('Food')['Likeness'].agg(['min','max']))

```

```python
# Use Groupby of any 2 columns with aggregate mean()
data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'Sales', 'Sales'],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F'],
    'Salary': [50000, 60000, 40000, 45000, 55000, 50000]
}

df = pd.DataFrame(data)

print(df.groupby(['Department', 'Gender'])['Salary'].mean())

```

```python
# Use Groupby of any 2 columns with aggregate min() and max()

print(df.groupby(['Department', 'Gender'])['Salary'].agg(['min', 'max']))

```
